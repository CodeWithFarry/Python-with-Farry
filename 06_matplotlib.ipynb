# MATPLOTLIB TUTORIAL

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x = np.arange(0, 10)
y = np.arange(10,20)
```

```python
plt.scatter(x,y, c="g")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Graph of x and y')
```

```python
#plt.savefig("Test.png")
```

```python
y = x*x
```

```python
plt.plot(x,y,'ro--')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plot Graph 2D')
```

```python
plt.plot(x,y, 'r*--', linewidth= 2, markersize= 12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plot Graph 2D')
```

```python
# creating subplots
plt.subplot(2,2,1)
plt.plot(x,y, 'r--')
plt.subplot(2,2,2)
plt.plot(x,y, 'g*-')
plt.subplot(2,2,3)
plt.plot(x,y, 'bo')
```

```python
# ceating bar plots but seaborn creates better than matplotlib
x = [1,3,5]
y= [6,7,8]

x2 = [5,7,9]
y2= [4,6,8]
plt.bar(x,y)
plt.bar(x2, y2)  # this makes the second plot start below the first one
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Bar Plot 2D')
```

```python
# Histograms
a = np.arange(1, 70, step = 3)
plt.hist(a)
plt.title('histogram')
plt.show()
```

```python
# Boxplot using Matplotlib
data = [np.random.normal(0, std, 100) for std in range (1, 4)]
plt.boxplot(data, vert= True, patch_artist= True)
```

```python
# pie chart
labels = 'python', 'java', 'c++', 'streamlit'
sizes = 121, 565, 300, 455
colors = ['green', 'blue', 'red', 'yellow']
explode = (0.4, 0.01, 0.01, 0.01)

plt.pie(sizes, explode= explode, labels= labels ,  colors = colors,  autopct='%1.1f%%', shadow=True)
```
